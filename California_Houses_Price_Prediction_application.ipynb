```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing california Houses Price Dataset

```python
from sklearn.datasets import fetch_california_housing
house_price_dataset = fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
#loading dataset to dataframe
df = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
df.head()
```

```python
#add price column
df['price'] = house_price_dataset.target
df.head()
```

```python
#check shape
df.shape
```

```python
#Check for missing values
df.isnull().sum()
```

```python
# statistical measures of the dataset
df.describe()
```

Understanding the correlation between features

```python
corre = df.corr()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corre,  cbar=True, square=True, fmt='.1f' , annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
#Splitting dataset
X = df.drop(['price'], axis=1)
Y = df['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

evaluate

```python
#prediction on training data
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
#visiualize
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
